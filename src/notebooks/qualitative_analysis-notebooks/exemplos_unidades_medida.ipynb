```python
def process_un_medida(un_medida):
    un_medida_process = unidecode(str(un_medida).lower())

    unidade = ['un', 'unid', 'und', 'uni', 'unidad', 'unid.', 'unitario', 'unidade.',
                'un.', 'uni.', 'und.', 'unida', 'unita', 'und,', 'un1', 'unid..', 'um',
                '0000000000000000000000000000000000000000000unidade', 'u']
    unidades = ['unidade(s)', 'unidade_s', 'unidade _s', 'unids']
    hora = ['hr', 'hrs', 'horas']
    jogo = ['jg', 'jgs']
    peca = ['pe', 'pc', 'pc.']
    saco = ['sc', 'sacos']
    caixa = ['cx', 'cx.', '000000000000000000000000000000000000000000000caixa', 'cxa']
    comprimido = ['cp', 'comp', 'comprimido 1 unidade', 'comp.', 'cmp', 'compr',
                  'comprimid', 'com', 'cpr', 'capsula', 'cap',
                  '0000000000000000000000000000000000000000comprimido']
    pacote = ['pct', 'pt', 'pcte', 'pacote.', 'pct.', 'pacot']
    kg = ['kilo', 'quilo', 'quilograma', 'kilograma', 'kilos', 'quilos', 'quilogramas',
          'kilogramas', 'quilo.', 'quilograma_s']
    frasco = ['fr', 'frasco.', 'fras', 'frasco 1 unidade', 'frasco/ampola', 'frasc']
    litro = ['litros', 'ltr', 'lit', 'litro.', 'li', 'l', 'lt', 'lts']

    if un_medida_process in unidade:
        return 'unidade'
    if un_medida_process in unidades:
        return 'unidades'
    if un_medida_process in hora:
        return 'hora'
    if un_medida_process in jogo:
        return 'jogo'
    if un_medida_process in peca:
        return 'peca'
    if un_medida_process in saco:
        return 'saco'
    if un_medida_process in caixa:
        return 'caixa'
    if un_medida_process in comprimido:
        return 'comprimido'
    if un_medida_process in pacote:
        return 'pacote'
    if un_medida_process in kg:
        return 'kg'
    if un_medida_process in frasco:
        return 'frasco'
    if un_medida_process in litro:
        return 'litro'
    unidades_medida_sem_ref.append(un_medida_process)
    return un_medida_process
```

```python
def dist_freq_len(frequencias, title, xlabel):
    y= list(frequencias.values())
    x= list(frequencias.keys())
    fig, ax = plt.subplots(figsize=(8,8))
    ax.bar(x,y)
    ax.set_xticks(x)
    ax.set_xticklabels(x)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Frequência')
    plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
    plt.show()
```

```python
def process_apenas_numeros(un_dict, un_medida):
    if str(un_medida).isnumeric():
        un_dict[un_medida] += 1
```

```python
def process_com_numero(un_dict, un_medida):
    if any(c.isdigit() for c in str(un_medida)):
        un_dict[un_medida] += 1
```

```python
def process_com_preposicao(un_dict, prep_dict, un_medida):
    tokens = str(un_medida).split(' ')
    for token in tokens:
        if token.lower() in list_preposicao:
            prep_dict[token.lower()] += 1
            un_dict[un_medida] += 1
```

```python
def process_qtd_palavras(un_dict, un_medida):
    qtd_palavras = len(str(un_medida).split(' '))
    un_dict[un_medida] = (un_dict[un_medida][0] + 1, qtd_palavras)
```

```python
def process_qtd_caracteres(un_dict, un_medida):
    qtd_caracteres = len(str(un_medida))
    un_dict[un_medida] = (un_dict[un_medida][0] + 1, qtd_caracteres)
```

```python
def process_numero_esq(un_dict, un_medida):
    if str(un_medida[0])[0].isnumeric():
        un_dict[un_medida] += 1
```

```python
def process_numero_ext(un_dict, un_medida):
    for token in un_medida.split(' '):
        if token in lista_numeros_ext:
            un_dict[un_medida] += 1
```

```python
def print_tuple_data(data, title, reverse=True, orderBy=0, limit=30, printSize = 50, min_freq = 0, min_size = 0):
    data_list = sorted(data.items(), key=lambda x: x[1][orderBy], reverse=reverse)
    data_list = list(filter(lambda e: e[1][1] >= min_size, data_list))
    print(f'### {title} ###')
    output = ''
    for i , t in enumerate(data_list[:min(limit, len(data))]):
        if(t[1][0] >= min_freq and t[1][1] >= min_size):
            output += str(f'{i + 1}) {t[0]}: {t[1]}\n')
#         output += str(info + ' '*(printSize - len(info)))
#         if i % 3 == 0:
#             output += '\n'
    print(output)
```

```python
def print_data(data, title, limit = 30, printSize = 50, min_freq = 0):
    data_list = sorted(data.items(), key=lambda x: x[1], reverse=True)
    print(f'### {title} ###')
    output = ''
    for i,t in enumerate( data_list[:min(limit, len(data))]):
        output += str(f'{i + 1}) {t[0]}: {t[1]}\n')
#         output += str(info + ' '*(printSize - len(info)))
#         if i % 3 == 0:
#             output += '\n'
    print(output)
#     print('\n')
```

```python
apenas_numeros = defaultdict(int)
com_numeros = defaultdict(int)
numero_esq = defaultdict(int)
com_numero_extenso = defaultdict(int)
com_preposicao = defaultdict(int)
preposicoes = defaultdict(int)
qtd_palavras = defaultdict(lambda: (0, 0))
qtd_caracteres = defaultdict(lambda: (0, 0))


for _, row in df_druid_process.iterrows():
    un_medida = row['dsc_unidade_medida']
    process_apenas_numeros(apenas_numeros, un_medida)
#     un_medida = un_medida.lstrip('0123456789.- ')
    process_com_numero(com_numeros, un_medida)
    process_numero_esq(numero_esq, un_medida)
    process_numero_ext(com_numero_extenso, un_medida)
    process_com_preposicao(com_preposicao, preposicoes, un_medida)
    process_qtd_palavras(qtd_palavras, un_medida)
    process_qtd_caracteres(qtd_caracteres, un_medida)
```
