```python
from urllib.request import *
import re
from bs4 import BeautifulSoup
import hashlib
import os
from IPython.display import display
import pandas as pd
import glob
import xlrd
import csv
from os import sys
from datetime import datetime
str(datetime.now())
```

```python
pages = ["https://www.gov.br/anvisa/pt-br/assuntos/medicamentos/cmed/precos/anos-anteriores/anos-anteriores"]
```

```python
base_path = "../data/docs_medicamentos_cmed_anvisa"

for page in pages:
    content = str(urlopen(page).read())
    section_2 = content[re.search("2\) ", content).start():]
    soup = BeautifulSoup(section_2, 'html.parser')
    for a in soup.select("a"):
        span = a.select("span")
        if (span and span[0].text == "XLS") or (a and a.text == "XLS"):
            document_url = a.attrs["href"].replace("/view", "")
            response = urlopen(document_url)
            print("url: " + response.url)
            retries = 0
            while retries < 5:
                try:
                    data = response.read()
                    filename = (response.url.split("/")[-1].replace(".xls", "") + "_" + str(datetime.now()) + ".xls").replace(" ", "")
                    print("saving " + filename + "...")
                    with open(base_path + f"/{filename}", "wb") as f:
                        f.write(data)       
                    break
                except Exception as e:
                    print("retry: " + response.url, e)
                    response = urlopen(document_url)
                    retries += 1
```

```python
def calc_checksum(xls_docs):
    checksums = []
    for doc_name in xls_docs:
        with open(doc_name, "rb") as f:
            h = hashlib.md5(f.read()).hexdigest()
            checksums.append({"filename": doc_name, "md5hash": h})
    return checksums
```

```python
base_path = "../data/docs_medicamentos_cmed_anvisa/"
our_xls_docs = list(filter(lambda s: s.endswith(".xls"), os.listdir(base_path)))
our_xls_docs = [base_path + xls for xls in our_xls_docs]
our_xls_checksums = calc_checksum(our_xls_docs)
```

```python
crawler_checksums_df = pd.DataFrame(data=crawler_checksums)
our_xls_checksums_df = pd.DataFrame(data=our_xls_checksums)
df = crawler_checksums_df.merge(our_xls_checksums_df, on="md5hash", how="inner", suffixes=("_their", "_ours"))
print("Matches")
display(df)
print("Quantidade de matches: " + str(df.shape[0]))
```

```python
base_path = "../data/docs_medicamentos_cmed_anvisa/"

files = glob.glob(base_path + "*.xls")

for excel_file in files:
    try:
        workbook = xlrd.open_workbook(excel_file)
        all_worksheets = workbook.sheet_names()
        for worksheet_name in all_worksheets[:1]:
            worksheet = workbook.sheet_by_name(worksheet_name)
            with open(u'{}.csv'.format(excel_file.rstrip(".xls")), 'w', encoding="utf-8") as your_csv_file:
                wr = csv.writer(your_csv_file, quoting=csv.QUOTE_ALL)
                for rownum in range(worksheet.nrows):
                    wr.writerow(worksheet.row_values(rownum))
    except Exception as e:
        print("Error in " + excel_file, e)
```
