```python
count_item_no_group = 0
count_item_no_un_med = 0
count_item_proble_mais = 0
count_item_proble_menos = 0
count_item_no_min_qtd = 0
cont = 0
for index, item_result in items_clusters_test.iterrows():
    cont += 1
    if (cont%100000)==0:
        print(cont,items_clusters_test.shape[0])
    if item_result['cluster'] == '-2' :
        count_item_no_group += 1
        continue
    if item_result['mean'] == -1:
        count_item_no_un_med += 1
    else:
        if item_result['count']>=20:
            if item_result['price'] > (item_result['mean'] + 2*item_result['std']):
                count_item_proble_mais += 1
            if item_result['price'] < (item_result['mean'] - 2*item_result['std']):
                count_item_proble_menos += 1
        else:
          count_item_no_min_qtd += 1  

count_item_no_group, count_item_no_un_med, count_item_proble_mais, count_item_proble_menos,count_item_no_min_qtd
```

```python
def show_clusters_by_cluster_dsc(id_cluster, dsc_unidade_medida, dsc1, dsc2):
    selected = items_clusters_train
    len_group = 0
    if id_cluster:
        selected = items_clusters_train[items_clusters_train['cluster'].str.contains(id_cluster, na=False)]
        len_group = len(selected['cluster'].unique())
    if dsc1:
        selected = selected[selected['description'].str.contains(dsc1, na=False)]
    if dsc2:
        selected = selected[selected['description'].str.contains(dsc2, na=False)]
    if dsc_unidade_medida:
        selected = selected[selected['dsc_unidade_medida'].str.upper().str.contains(dsc_unidade_medida.upper(), na=False)]
        
    for idx, row in selected.iterrows():
        print('item_id:'+str(row['item_id']), 'description:'+row['description'], 'cluster:'+row['cluster'], 'dsc_unidade_medida:'+row['dsc_unidade_medida'])
```
